```python
#Purchasing Product


prod  = str(input("Enter the product_Id: "))
quant = int(input("Enter the quantity: "))
if(data[prod]['pr']>quant):
    print("Product: ", data[prod]['name'])
    print("Price: ", data[prod]['pr'])
    print("Total Bill: ", data[prod]['pr'] * quant)

    data[prod]['qn'] = data[prod]['qn'] - quant
else:
    print("Input quantity is greater than the available quantity")
```

```python
js = json.dumps(data)

fd = open("record.json",'w')
fd.write(js)
fd.close()
```
